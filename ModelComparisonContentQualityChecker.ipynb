````python
import ast
import json
import re
import time

if client is None or not ListOfModels:
    raise RuntimeError("Cell 1 is not configured yet. Set HF_TOKEN before running evaluations.")

# Optional explicit alias map (left empty by default).
MODEL_ALIASES: dict[str, str] = {}


def is_model_not_found(error_text: str) -> bool:
    txt = (error_text or "").lower()
    return ("model_not_found" in txt) or ("does not exist" in txt and "model" in txt)


def _unique_keep_order(values: list[str]) -> list[str]:
    seen = set()
    out: list[str] = []
    for v in values:
        if v and v not in seen:
            seen.add(v)
            out.append(v)
    return out


def model_candidates(model_name: str, alias_map: dict[str, str] | None = None) -> list[str]:
    """Generate candidate model IDs to recover from provider/casing mismatches."""
    candidates: list[str] = [model_name]
    aliases = alias_map or MODEL_ALIASES

    # Case-insensitive alias lookup
    for k, v in aliases.items():
        if model_name.lower() == k.lower():
            candidates.append(v)

    # Base model without provider suffix
    if ":" in model_name:
        base = model_name.split(":", 1)[0]
        candidates.append(base)
    else:
        # Try common providers if missing suffix
        for suffix in (":groq", ":novita", ":featherless-ai"):
            candidates.append(f"{model_name}{suffix}")

    return _unique_keep_order(candidates)


def evaluate_story(model_name: str, story: str, rubric: str) -> dict:
    """Send a story to a model for quality evaluation and return structured results."""
    weirdness_guard = """
Additional Quality Gate (Weirdness Control):
- Penalize scenes that feel random, absurd, or unintentionally bizarre for ages 8-12.
- Penalize exaggerated reactions that break immersion unless clearly justified by context.
- Penalize non-sequiturs, chaotic tonal swings, or gimmicky effects that distract from plot and learning.
- Reward clear cause-effect logic, grounded humor, and emotionally believable character behavior.
- If weirdness reduces clarity or educational value, deduct points mainly in narrative_flow, tone_vocabulary, and character_agency.
""".strip()
    prompt = rubric.replace("{StoryDraft}", story)
    prompt = f"{prompt}\n\n{weirdness_guard}"
    result = {
        "model": model_name,
        "model_used": None,
        "response": None,
        "story_category": None,
        "final_score": None,
        "scores": {},
        "error": None,
        "latency_seconds": None,
    }

    last_error = None
    for candidate in model_candidates(model_name):
        try:
            start = time.time()
            completion = client.chat.completions.create(
                model=candidate,
                messages=[
                    {
                        "role": "system",
                        "content": "You are a precise evaluator that must follow the requested output schema exactly. Return JSON object only.",
                    },
                    {"role": "user", "content": prompt},
                ],
                max_tokens=1500,
                temperature=0,
            )
            result["latency_seconds"] = round(time.time() - start, 2)
            result["response"] = completion.choices[0].message.content
            result["model_used"] = candidate
            return result
        except Exception as e:
            err = str(e)
            last_error = err
            if is_model_not_found(err):
                continue
            result["error"] = err
            return result

    result["error"] = last_error or "Model resolution failed."
    return result


def _to_int(value):
    if value is None:
        return None
    if isinstance(value, bool):
        return int(value)
    if isinstance(value, (int, float)):
        return int(round(float(value)))
    if isinstance(value, str):
        m = re.search(r"-?\d+(?:\.\d+)?", value)
        if m:
            return int(round(float(m.group(0))))
    return None


def _find_number_by_key(text: str, key: str) -> int | None:
    if not text:
        return None
    patterns = [
        rf'"{re.escape(key)}"\s*:\s*"?(-?\d+(?:\.\d+)?)"?',
        rf"'{re.escape(key)}'\s*:\s*'?(-?\d+(?:\.\d+)?)'?",
        rf"\b{re.escape(key)}\b\s*[:=]\s*" + r'"?(-?\d+(?:\.\d+)?)"?',
    ]
    for p in patterns:
        m = re.search(p, text, re.IGNORECASE)
        if m:
            return _to_int(m.group(1))
    return None


def extract_json_object(response_text: str) -> dict | None:
    """Best-effort extraction of a JSON object from raw model output."""
    if not response_text:
        return None

    cleaned = response_text.strip()

    # Remove markdown code fences when present
    if cleaned.startswith("```"):
        cleaned = re.sub(r"^```(?:json)?\s*", "", cleaned, flags=re.IGNORECASE)
        cleaned = re.sub(r"\s*```$", "", cleaned)

    # Direct parse first
    try:
        payload = json.loads(cleaned)
        if isinstance(payload, dict):
            return payload
    except json.JSONDecodeError:
        pass

    # Fallback: extract largest brace block and try JSON then Python literal parsing
    start = cleaned.find("{")
    end = cleaned.rfind("}")
    if start != -1 and end != -1 and end > start:
        candidate = cleaned[start : end + 1]

        try:
            payload = json.loads(candidate)
            if isinstance(payload, dict):
                return payload
        except json.JSONDecodeError:
            pass

        py_candidate = re.sub(r"\bnull\b", "None", candidate, flags=re.IGNORECASE)
        py_candidate = re.sub(r"\btrue\b", "True", py_candidate, flags=re.IGNORECASE)
        py_candidate = re.sub(r"\bfalse\b", "False", py_candidate, flags=re.IGNORECASE)
        try:
            payload = ast.literal_eval(py_candidate)
            if isinstance(payload, dict):
                return payload
        except (ValueError, SyntaxError):
            pass

    return None


def parse_score(response_text: str) -> dict:
    """Extract numerical scores from model response, preferring JSON output."""
    parsed = {
        "story_category": None,
        "final_score": None,
        "narrative_flow": None,
        "educational_integration": None,
        "scientific_accuracy": None,
        "tone_vocabulary": None,
        "read_aloud": None,
        "character_agency": None,
    }

    payload = extract_json_object(response_text)
    if isinstance(payload, dict):
        breakdown = payload.get("score_breakdown")
        if not isinstance(breakdown, dict):
            breakdown = payload.get("scores") if isinstance(payload.get("scores"), dict) else {}

        parsed["story_category"] = payload.get("story_category") or payload.get("category")
        parsed["final_score"] = _to_int(payload.get("final_score") or payload.get("overall_score") or payload.get("total_score"))
        parsed["narrative_flow"] = _to_int(breakdown.get("narrative_flow"))
        parsed["educational_integration"] = _to_int(breakdown.get("educational_integration"))
        parsed["scientific_accuracy"] = _to_int(breakdown.get("scientific_accuracy"))
        parsed["tone_vocabulary"] = _to_int(breakdown.get("tone_vocabulary"))
        parsed["read_aloud"] = _to_int(breakdown.get("read_aloud"))
        parsed["character_agency"] = _to_int(breakdown.get("character_agency"))

    # If still missing values, use resilient text extraction by canonical keys.
    key_map = {
        "final_score": ["final_score", "overall_score", "total_score", "final score", "overall score", "total score"],
        "narrative_flow": ["narrative_flow", "narrative flow"],
        "educational_integration": ["educational_integration", "educational integration"],
        "scientific_accuracy": ["scientific_accuracy", "scientific accuracy"],
        "tone_vocabulary": ["tone_vocabulary", "tone and vocabulary", "age-appropriate tone", "tone"],
        "read_aloud": ["read_aloud", "read-aloud", "read aloud"],
        "character_agency": ["character_agency", "character action and agency", "character agency"],
    }

    for field, aliases in key_map.items():
        if parsed[field] is None:
            for alias in aliases:
                val = _find_number_by_key(response_text or "", alias)
                if val is not None:
                    parsed[field] = val
                    break

    # Final fallback to legacy regex labels
    if parsed["final_score"] is None and response_text:
        legacy_patterns = {
            "final_score": r"(?:Final Score|Overall Score|Total Score)[:\s]+(\d+(?:\.\d+)?)\s*(?:/\s*100)?",
            "narrative_flow": r"Narrative Flow.*?[:\s]+(\d+(?:\.\d+)?)\s*(?:/\s*20)?",
            "educational_integration": r"Educational Integration.*?[:\s]+(\d+(?:\.\d+)?)\s*(?:/\s*20)?",
            "scientific_accuracy": r"Scientific Accuracy.*?[:\s]+(\d+(?:\.\d+)?)\s*(?:/\s*20)?",
            "tone_vocabulary": r"(?:Age.?Appropriate Tone|Tone and Vocabulary).*?[:\s]+(\d+(?:\.\d+)?)\s*(?:/\s*15)?",
            "read_aloud": r"Read.?Aloud.*?[:\s]+(\d+(?:\.\d+)?)\s*(?:/\s*10)?",
            "character_agency": r"(?:Character Action and Agency|Character Agency).*?[:\s]+(\d+(?:\.\d+)?)\s*(?:/\s*15)?",
        }
        for key, pattern in legacy_patterns.items():
            if parsed[key] is None:
                match = re.search(pattern, response_text, re.IGNORECASE)
                if match:
                    parsed[key] = _to_int(match.group(1))

    # If final score is missing but all category scores are available, compute it.
    component_keys = [
        "narrative_flow",
        "educational_integration",
        "scientific_accuracy",
        "tone_vocabulary",
        "read_aloud",
        "character_agency",
    ]
    if parsed["final_score"] is None and all(parsed[k] is not None for k in component_keys):
        parsed["final_score"] = int(sum(parsed[k] for k in component_keys if parsed[k] is not None))

    return parsed


def normalize_story_drafts() -> list[dict]:
    """Support both old single/multi StoryDraft and new StoryDrafts formats."""
    if "StoryDrafts" in globals() and isinstance(StoryDrafts, list) and StoryDrafts:
        if isinstance(StoryDrafts[0], dict):
            return StoryDrafts
        if isinstance(StoryDrafts[0], str):
            return [
                {
                    "name": f"Draft {i + 1}",
                    "category": globals().get("StoryCategory", "Children's educational fiction"),
                    "text": text,
                }
                for i, text in enumerate(StoryDrafts)
            ]

    if "StoryDraft" in globals():
        if isinstance(StoryDraft, list):
            return [
                {
                    "name": f"Draft {i + 1}",
                    "category": globals().get("StoryCategory", "Children's educational fiction"),
                    "text": text,
                }
                for i, text in enumerate(StoryDraft)
            ]
        if isinstance(StoryDraft, str):
            return [
                {
                    "name": "Draft 1",
                    "category": globals().get("StoryCategory", "Children's educational fiction"),
                    "text": StoryDraft,
                }
            ]

    raise RuntimeError("No story drafts found. Define StoryDraft (str/list) or StoryDrafts (list).")


def normalize_instruction_variants(default_story_category: str) -> list[dict]:
    """Support both old single instruction and new InstructionVariants format."""
    if "InstructionVariants" in globals() and isinstance(InstructionVariants, list) and InstructionVariants:
        if "build_instruction" in globals() and callable(build_instruction):
            variants = []
            for i, variant in enumerate(InstructionVariants):
                name = variant.get("name", f"Instruction {i + 1}")
                rubric = build_instruction(variant, default_story_category)
                variants.append({"name": name, "rubric": rubric})
            return variants
        if isinstance(InstructionVariants[0], dict) and "rubric" in InstructionVariants[0]:
            return InstructionVariants

    if "instruction" in globals() and isinstance(instruction, str):
        return [{"name": "Instruction 1", "rubric": instruction}]

    raise RuntimeError(
        "No instruction found. Define `instruction` or `InstructionVariants` + `build_instruction`."
    )


story_drafts = normalize_story_drafts()
instruction_variants = normalize_instruction_variants(story_drafts[0]["category"])

print(f"Models: {len(ListOfModels)} | Drafts: {len(story_drafts)} | Instructions: {len(instruction_variants)}")

# Run evaluations across models x drafts x instructions
results = []
for draft in story_drafts:
    for instr in instruction_variants:
        print(f"=== {draft['name']} | {instr['name']} ===")
        for model in ListOfModels:
            print(f"Evaluating: {model} ...")
            result = evaluate_story(model, draft["text"], instr["rubric"])
            if result["response"]:
                result["scores"] = parse_score(result["response"])
                result["story_category"] = result["scores"].get("story_category")
                result["final_score"] = result["scores"].get("final_score")

            result["draft_name"] = draft["name"]
            result["draft_category"] = draft.get("category")
            result["instruction_name"] = instr["name"]
            results.append(result)

            if result["final_score"] is not None:
                status = f"Score: {result['final_score']}/100"
            elif result["error"]:
                status = f"Error: {result['error']}"
            else:
                status = "Unparsed response"

            used = result.get("model_used") or model
            print(f"  -> {status} | Model used: {used} | Latency: {result['latency_seconds']}s")
        print()

print(f"All evaluations complete. Total runs: {len(results)}")
````

````python
# --- Comparison Tables ---
score_labels = {
    "final_score": "Final /100",
    "narrative_flow": "Narrative /20",
    "educational_integration": "Education /20",
    "scientific_accuracy": "Science /20",
    "tone_vocabulary": "Tone /15",
    "read_aloud": "Read-Aloud /10",
    "character_agency": "Agency /15",
}

col_width = 32
label_width = 16

# Table 1: every run
print("ALL RUNS")
header = (
    f"{'Draft':<24}{'Instruction':<28}{'Model':<{col_width}}"
    + "".join(f"{v:>{label_width}}" for v in score_labels.values())
)
print(header)
print("-" * len(header))

for r in sorted(
    results,
    key=lambda x: (x.get("draft_name", ""), x.get("instruction_name", ""), -(x.get("final_score") or 0)),
):
    short_model = r["model"].split("/")[-1][:30]
    row = f"{r.get('draft_name', 'N/A')[:23]:<24}{r.get('instruction_name', 'N/A')[:27]:<28}{short_model:<{col_width}}"
    for key in score_labels:
        val = r.get("scores", {}).get(key) if r.get("scores") else None
        row += f"{str(val) if val is not None else 'N/A':>{label_width}}"
    print(row)

print()
print("BEST MODEL PER DRAFT x INSTRUCTION")
best_map = {}
for r in results:
    k = (r.get("draft_name"), r.get("instruction_name"))
    if k not in best_map or (r.get("final_score") or -1) > (best_map[k].get("final_score") or -1):
        best_map[k] = r

for (draft_name, instr_name), r in sorted(best_map.items()):
    print(
        f"- {draft_name} | {instr_name} -> {r['model']} | "
        f"Score: {r.get('final_score', 'N/A')} | Latency: {r.get('latency_seconds', 'N/A')}s"
    )

print()
print("BEST INSTRUCTION PER DRAFT (average over models)")
agg = {}
for r in results:
    score = r.get("final_score")
    if score is None:
        continue
    k = (r.get("draft_name"), r.get("instruction_name"))
    agg.setdefault(k, []).append(score)

best_instruction_by_draft = {}
for (draft_name, instr_name), vals in agg.items():
    avg_score = round(sum(vals) / len(vals), 2)
    rec = (instr_name, avg_score, len(vals))
    if draft_name not in best_instruction_by_draft or avg_score > best_instruction_by_draft[draft_name][1]:
        best_instruction_by_draft[draft_name] = rec

for draft_name, (instr_name, avg_score, n) in sorted(best_instruction_by_draft.items()):
    print(f"- {draft_name}: {instr_name} | Avg Score: {avg_score} (n={n} models)")
````
